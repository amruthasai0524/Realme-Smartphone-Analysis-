```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import time

productname = []
price = []
rating = []
features = []
pagenum = []
original_price = []
Discount = []
Review = []
total_time = time.time()

for i in range(1,414):
    start_time = time.time()
    URL = f"https://www.flipkart.com/search?q=realme+5g+mobile&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_3_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_3_6_na_na_na&as-pos=3&as-type=HISTORY&suggestionId=realme+5g+mobile%7CMobiles&requestId=9c320fe5-d18f-40f2-b35b-680e2ea1c831&page={i}"
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")

    for x in soup.find_all('div', attrs = {'class':'jIjQ8S'}):
        pname = x.find('div', attrs={'class':'RG5Slk'})
        pri = x.find('div', attrs={'class':'hZ3P6w DeU9vF'})  
        rat = x.find('div', attrs={'class':'MKiFS6'})
        feat = x.find('div', attrs={'class':'CMXw7N'})
        op = x.find('div', attrs={'class':'kRYCnD gxR4EY'})
        discountp = x.find('div', attrs={'class':'HQe8jr'})
        reviews = x.find('span', attrs={'class':'PvbNMB'})

        
        productname.append(pname.text if pname else np.nan)
        price.append(pri.text if pri else np.nan)
        rating.append(rat.text if rat else np.nan)
        features.append(feat.text if feat else np.nan)
        original_price.append(op.text if op else np.nan)
        Discount.append(discountp.text if discountp else np.nan)
        Review.append(reviews.text if reviews else np.nan)
        
        pagenum.append(i)
    
```

```python
import pandas as pd
```

```python
data = pd.DataFrame({"product name":productname,"price":price,"ratings":rating,"features":features,"page number":pagenum,"original_price":original_price,"Discount":Discount,"Review":Review})
```
